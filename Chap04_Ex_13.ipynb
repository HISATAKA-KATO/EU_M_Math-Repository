```python
#今回利用するライブラリ
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

#可視化ライブラリ
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
%matplotlib inline

#小数点第三位まで表示
%precision 3

#ランダムシード
np.random.seed(0)

#zipダウンロードライブラリ
import requests, zipfile
from io import StringIO
import io
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'

r = requests.get(url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()


student_data_math = pd.read_csv('student-mat.csv',sep = ';')

student_data_por = pd.read_csv('student-por.csv',sep = ';')

student_data_merge = pd.merge(student_data_math
                              ,student_data_por
                              ,on = ['school','sex','age','address','famsize','Pstatus','Medu'
                              ,'Fedu','Mjob','Fjob','reason','nursery','internet']
                              ,suffixes = ('_math','_por'))

print('G1数学の成績平均：', student_data_merge.G1_math.mean())
print('G1ポルトガル語の成績平均：',student_data_merge.G1_por.mean())

print('\nG2数学の成績平均：', student_data_merge.G2_math.mean())
print('G2ポルトガル語の成績平均：',student_data_merge.G2_por.mean())

print('\nG3数学の成績平均：', student_data_merge.G3_math.mean())
print('G3ポルトガル語の成績平均：',student_data_merge.G3_por.mean())

#検定
from scipy import stats
t,p1 = stats.ttest_rel(student_data_merge.G1_math,student_data_merge.G1_por)
print('\np値(G1) = ',p1)

t,p2 = stats.ttest_rel(student_data_merge.G2_math,student_data_merge.G2_por)
print('\np値(G2) = ',p2)

t,p3 = stats.ttest_rel(student_data_merge.G3_math,student_data_merge.G3_por)
print('\np値(G3) = ',p3)
```
