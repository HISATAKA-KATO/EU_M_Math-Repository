```python
#今回利用するライブラリ
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

#可視化ライブラリ
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
%matplotlib inline

#zipダウンロードライブラリ
import requests, zipfile
from io import StringIO
import io
```

```python
import urllib.request

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"  # ダウンロードしたいURL
save_name = "winequality-red.csv"
urllib.request.urlretrieve(url, save_name)
print("ダウンロードが完了しました")

wine_qualty_data = pd.read_csv('winequality-red.csv',sep = ';')

wine_qualty_data.head()
```

```python
#ヒストグラム。データ指定
plt.plot(wine_qualty_data['residual sugar'],wine_qualty_data['free sulfur dioxide'],'o')
''
#x軸、y軸のそれぞれのラベル
plt.xlabel('residual sugar')
plt.ylabel('free sulfur dioxide')
plt.grid(True)

#平均値(mean())
print('平均値：',wine_qualty_data['residual sugar'].mean())

#中央値(median())
print('中央値：',wine_qualty_data['residual sugar'].median())

#最頻値(var())
print('最頻値：',wine_qualty_data['residual sugar'].mode())

#分散(var())
print('分散：',wine_qualty_data['residual sugar'].var())

#標準偏差(std())
print('標準偏差：',wine_qualty_data['residual sugar'].std())
```
