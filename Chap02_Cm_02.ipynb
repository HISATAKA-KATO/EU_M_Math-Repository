```python
# 以下にライブラリを定義する。
import numpy as np
import numpy.random as random
import scipy as sp
import scipy.linalg  as linalg  ##行列式用
from scipy.optimize import newton ##ニュートン法
import pandas as pd
from pandas import Series, DataFrame


# 可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

# 少数第3位まで表示
%precision 3
```

```python
random.seed(0)

x = np.random.uniform(0.0,1.0,10000)
y = np.random.uniform(0.0,1.0,10000)

#円の内部判定
inside = (x**2 + y**2 <= 1)

# 円の内側
plt.scatter(x[inside], y[inside], s=5, label='Inside')

# 円の外側
plt.scatter(x[~inside], y[~inside], s=5, label='Outside')

# 円を描画
theta = np.linspace(0, 2*np.pi, 500)
plt.plot(np.cos(theta), np.sin(theta))

# 軸をそろえる
plt.axis('equal')

# 範囲
plt.xlim(0, 1)
plt.ylim(0, 1)

# タイトル
plt.title("Random Points and Unit Circle")

# 表示
plt.show()

# 円内部の点数
print("円内部の点数:", np.sum(inside))
```
