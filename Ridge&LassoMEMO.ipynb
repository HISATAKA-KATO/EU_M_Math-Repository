リッジ回帰では、係数の二乗和を正則化項として用いている。それによって、係数が完全に0にならず、全体的に小さくなるため、多重共線性がある場合でも安定した推定ができる。また、ラッソ回帰では、係数の絶対値の和を正則化項として用いるため、不要な説明変数の係数を0にできるため重要な変数だけを選ぶ「特徴量選択」が可能である。
