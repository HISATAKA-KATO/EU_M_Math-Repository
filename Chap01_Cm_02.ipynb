```python
N = int(input("自然数"))

number = 1

while number <= N:
    if number >= 2:
        divisor = 2
        prime_number = True

        while divisor < number:
            if number % divisor == 0:
                prime_number = False
            divisor = divisor + 1

        if prime_number:
            print(number)

    number = number + 1
```
