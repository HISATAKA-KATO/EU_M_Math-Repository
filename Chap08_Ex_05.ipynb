```python
#標準化ライブラリ
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier#決定木モデル

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# データの読み込み
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)

# デフォルトの決定木モデルの構築
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)

print(f"★デフォルト★")
print(f"訓練セットのスコア: {tree.score(X_train, y_train):.3f}")
print(f"テストセットのスコア: {tree.score(X_test, y_test):.3f}")
```

```python
# パラメータを変更したモデル（木の深さを制限）
tree_trans = DecisionTreeClassifier(max_depth=4, random_state=0)
tree_trans.fit(X_train, y_train)

print(f"★変更後★")
print(f"訓練セットのスコア: {tree_trans.score(X_train, y_train):.3f}")
print(f"テストセットのスコア: {tree_trans.score(X_test, y_test):.3f}")
```
