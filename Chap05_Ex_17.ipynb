```python
from scipy.optimize import newton

# 関数の定義
def f(x):
    return 5 * x - 10

# ニュートン法を用いて解を求める（初期値 x=0 と設定）
root = newton(f, 0)

print(f"f(x) = 5x - 10 が 0 となる x は: {root}")
```
