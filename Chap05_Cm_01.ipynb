```python
A = np.array([[5,1,0,1],
              [1,9,-5,7],
              [0,-5,8,-3],
              [1,7,-3,10]])
b = np.array([2,10,5,10])

L = sp.linalg.cholesky(A)
t = sp.linalg.solve(L.T.conj(),b)
x = sp.linalg.solve(L,t)

print(x)
```
