```python
import numpy as np
import matplotlib.pyplot as plt
import random

random.seed(0)

# 点の数
N = 10000

# 一様乱数
x = np.random.uniform(0.0, 1.0, N)
y = np.random.uniform(0.0, 1.0, N)

# 円内部判定
inside = (x**2 + y**2 <= 1)

# 内部点数
inside_count = np.sum(inside)

# 円周率近似
pi_estimate = 4 * inside_count / N

print("円内部の点数:", inside_count)
print("円周率近似:", pi_estimate)

# プロット
plt.scatter(x[inside], y[inside], s=5, label='Inside')
plt.scatter(x[~inside], y[~inside], s=5, label='Outside')

# 円弧
theta = np.linspace(0, np.pi/2, 300)
plt.plot(np.cos(theta), np.sin(theta))

plt.axis('equal')
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.legend()
plt.show()
```
