```python
#ｘが0以上1以下,ｙは0以上yで1-x以下の三角領域で定義される以下の関数の積分値を求めたい。
def f(y,x):
  return 1/(np.sqrt(x + y) * ((1 + x + y) ** 2))

integral, abserr = integrate.dblquad(f, 0, 1, lambda x: 0, lambda x: 1 - x)
print(f"積分値: {integral}")
print(f"絶対誤差: {abserr}")
```
