```python
#最適化問題---min f(x) = x^2 + 1, s.t.x >= 1

from scipy.optimize import minimize

def f(x):
  y = x**2 + 1
  return y

initial = 0.0
constraint = [(-1, None)]

result_new = minimize(f, initial, bounds=constraint)

print("Optimization Result (x >= -1):")
print(result_new)

print(f"\nOptimized x: {result_new.x[0]:.3f}")
print(f"Optimized y: {result_new.fun:.3f}")
```
