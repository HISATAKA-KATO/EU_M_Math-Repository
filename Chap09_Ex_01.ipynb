```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

# random_state=52でデータを作成
X, _ = make_blobs(random_state=52)

# 元データの可視化
plt.scatter(X[:, 0], X[:, 1], color='blue')
plt.title('Generated Data (random_state=52)')
plt.show()
```

```python
import pandas as pd

kmeans = KMeans(init='k-means++', n_clusters=3, random_state=42)
kmeans.fit(X)
y_pred = kmeans.predict(X)

merge_data = pd.concat([pd.DataFrame(X[:, 0]), pd.DataFrame(X[:, 1]), pd.DataFrame(y_pred)], axis=1)
merge_data.columns = ['feature1', 'feature2', 'cluster']

ax = None
colors = ['blue', 'red', 'green']
for i, data in merge_data.groupby('cluster'):
    ax = data.plot.scatter(x='feature1', y='feature2', color=colors[i], label='cluster ' + str(i), ax=ax)
plt.title('Clustering Results')
```
