```python
#今回利用するライブラリ
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
import numpy.random as random
from scipy import interpolate

#可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set_style()
%matplotlib inline

#小数点第三位まで表示
%precision 3

#ランダムシード
np.random.seed(0)

#zipダウンロードライブラリ
import requests, zipfile
from io import StringIO
import io
```

```python
#データ
x = np.linspace(0,10,num = 11,endpoint = True)
y = np.sin(x**2/5.0)
xnew = np.linspace(0,10,num = 60,endpoint = True)

#線形補間
f1 = interpolate.interp1d(x,y,'linear')

#スプライン補間 (2次)
f2 = interpolate.interp1d(x,y,'quadratic')

#スプライン補間 (3次)
f3 = interpolate.interp1d(x,y,'cubic')

#グラフ描画
plt.plot(x,y,'o', label='data')
plt.plot(xnew, f1(xnew), '-', label='linear')
plt.plot(xnew, f2(xnew), '--', label='quadratic')#二次
plt.plot(xnew, f3(xnew), ':', label='cubic')#三次

plt.legend(loc = 'best')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
