#k-means
k-meansは、データを指定したクラスタ数 k 個に分ける代表的なクラスタリング手法である。各クラスタの中心を**重心（centroid）**として表し、各データ点を最も近い重心に割り当て、その後クラスタ内の平均値で重心を更新する。この処理を、重心の変化が小さくなるまで繰り返す。scikit-learnの説明でも、k-meansはクラスタ内平方和、つまり inertia を小さくすることを目的とするとされている。
\
**@メリット・デメリット**
\
メリットは、アルゴリズムが単純で理解しやすく、計算が速いため、大きなデータにも使いやすい点である。scikit-learnでも、k-meansは多くのサンプル数に対してスケールしやすい手法と説明されている。\
デメリットは、最初にクラスタ数 k を決める必要があること、初期値によって結果が変わりやすいこと、外れ値に弱いこと、球状でないクラスタや密度が異なるクラスタには弱いことである。また、重心は平均値なので、実際のデータ点とは限らない。

#k-means++
\
**k-means++**は、k-meansの改良版というより、k-meansの初期重心を賢く選ぶ方法である。通常のk-meansでは初期重心をランダムに選ぶが、k-means++では重心同士ができるだけ離れるように初期値を選ぶため、ランダム初期化よりも良い結果になりやすく、収束も速くなりやすい。scikit-learnでは、init="k-means++" がKMeansのデフォルトになっている。
\
**@メリット・デメリット**
\
メリットは、初期値の影響を減らせること、局所解に落ちるリスクをある程度下げられること、通常のランダム初期化より安定したクラスタリング結果が得られやすいことである。scikit-learnの説明でも、k-means++は初期重心を工夫して選ぶことで収束を速める手法とされている。\
デメリットは、あくまで初期値の工夫であり、k-meansそのものの弱点である「クラスタ数 k を事前に決める必要がある」「外れ値に弱い」「球状でないクラスタに弱い」といった問題は残る点である。

#k-medoids
\
k-medoidsは、k-meansに似たクラスタリング手法だが、クラスタの中心として平均値ではなく、実際のデータ点の中から代表点を選ぶ。この代表点をmedoidという。k-meansの中心はデータ上に存在しない場合があるが、k-medoidsの中心は必ず実在するデータ点になる。scikit-learn-extraのKMedoidsでは、クラスタ中心は元データ内の要素、つまりmedoidとして扱われる。\
**@メリット・デメリット**
\
メリットは、中心が実際のデータ点であるため解釈しやすいこと、平均値を使うk-meansより外れ値の影響を受けにくいこと、ユークリッド距離以外の距離尺度も使いやすいことである。\
デメリットは、k-meansより計算コストが大きくなりやすいことである。
