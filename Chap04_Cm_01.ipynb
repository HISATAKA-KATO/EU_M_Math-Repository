```python
#今回利用するライブラリ
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

#可視化ライブラリ
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
%matplotlib inline

#小数点第三位まで表示
%precision 3

#ランダムシード
np.random.seed(0)

#zipダウンロードライブラリ
import requests, zipfile
from io import StringIO
import io
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'

r = requests.get(url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()


student_data_math = pd.read_csv('student-mat.csv',sep = ';')

student_data_por = pd.read_csv('student-por.csv',sep = ';')

student_data_merge = pd.merge(student_data_math
                              ,student_data_por
                              ,on = ['school','sex','age','address','famsize','Pstatus','Medu'
                              ,'Fedu','Mjob','Fjob','reason','nursery','internet']
                              ,suffixes = ('_math','_por'))

print('\n数学の欠席数平均：', student_data_merge.absences_math.mean())
print('ポルトガル語の欠席数平均：', student_data_merge.absences_por.mean())
print('\n数学の勉強時間平均：', student_data_merge.studytime_math.mean())
print('ポルトガル語の勉強時間平均：', student_data_merge.studytime_por.mean())

# 欠席数に関する対応のある検定
from scipy import stats

t_abs, p_abs = stats.ttest_rel(student_data_merge.absences_math, student_data_merge.absences_por)
t_std, p_std = stats.ttest_rel(student_data_merge.studytime_math, student_data_merge.studytime_por)


print('\np値(欠席数) = ', p_abs)
print('\np値(勉強時間) = ', p_std)
```

以上より、欠席数に対して検定を実施した結果、P値は偶然性を示さなかった。一方、勉強時間に対して検定を実施した結果、P値は偶然性を示していた。従って、欠席数に関しては差があると言え、勉強時間に関しては差があると言えない。
