```python
num = 1

while num <= 10:
    if num >= 2:
        divisor = 2
        is_prime = True

        while divisor < num:
            if num % divisor == 0:
                is_prime = False
            divisor = divisor + 1

        if is_prime:
            print(num)

    num = num + 1
```
