```python
# 1. 必要なライブラリとIrisデータセットの読み込み
from pycaret.datasets import get_data
from pycaret.classification import *

# 講義資料の指示通り、irisデータを取得
data = get_data('iris')

# 2. 列名の確認
print("--- データセットの列名一覧 ---")
print(data.columns)

# 3. 環境のセットアップ（目的変数はアヤメの種類 'species'）
# データの自動前処理と学習用・テスト用の分割を行います
clf = setup(data, target='species', session_id=123)

# 4. すべての機械学習モデルを訓練・比較し、最も精度の高いモデルを選択
print("--- モデルの一斉比較を開始 ---")
best_model = compare_models()

# 5. 選択された最良モデルのハイパーパラメーターをチューニング
print("--- 最良モデルのチューニングを開始 ---")
tuned_model = tune_model(best_model)

# 6. テストデータを使ってモデルの最終評価（評価グラフ等の出力）
print("--- モデルの最終評価 ---")
evaluate_model(tuned_model)
```
