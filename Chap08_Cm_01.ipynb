#教師有学習の用語(1)(日常でどのように用いられるか、意味)
#用語

#・回帰
意味：いくつかのデータをもとに、「連続する具体的な値」を予測すること。
日常：日常的に行っている「予測」

#・分類
意味：いくつかのデータをもとに、「あらかじめ決まったグループ（カテゴリ・クラス）のどれに当てはまるか」を予測すること
日常：医師が「この状態なら健康（異常なし）」「この影は病気の疑いあり（異常あり）」と予測するなど

#・教師有り学習
意味：「問題（データ）」と「正解（ラベル）」のセットをコンピュータに大量に与えて、その特徴やルールを勉強させる学習法。
日常：幼い頃に言葉や知識を覚えたプロセス

#・重回帰分析
意味：「いくつかの原因（2つ以上のデータ）」をもとに、「1つの結果（数値）」を予測・分析する手法
日常：回帰と同じ。ただし、データ(説明変数)は2つ以上。

#・ロジスティック回帰分析
意味：いくつかの原因データをもとに、「ある出来事が起きる確率（0〜100%）」を計算し、最終的に「YESかNOか」を予測・分類する手法
日常：いくつかの条件から「確率」を頭の中で計算し、最終的にやるか・やらないかの2択を決めること

#・リッジ回帰
意味：重回帰分析の弱点である「過学習（オーバーフィッティング）」を防ぐために、データの「重み（影響度）」が大きくなりすぎないようブレーキ（ペナルティ）をかける手法
日常：「過去の極端な経験に囚われすぎず、少しマイルド（保守的）に見積もる」という考え

#・ラッソ回帰
意味：リッジ回帰と異なり、影響の薄い原因の重みを「完全にゼロにする」（なかったことにする）。
日常：「やることが多すぎてパニックになる前に、重要じゃないタスクを完全に切り捨てて、本質だけに集中する」という行動

#・決定木
意味：データの中から「YESかNOかで答えられる質問」を繰り返して、トーナメント表（樹形図）のような「木」の形を作りながらデータを予測・分類する手法

日常：「はい／いいえ」で進む心理テストや、フローチャート

#・エントロピー
意味：「そのデータがどれくらい乱雑で、予測しにくいか（不確実か）」を数値にしたもの
日常：「部屋の散らかり具合」や「結果のハラハラ度合」

#・情報利得
意味：「質問をする前のエントロピー」から「質問をした後のエントロピー」を引いた引き算の答え
日常：「相手の正体を突き止めるための、鋭い質問」

#・k-NN法
意味：新しいデータを分類するときに、「そのデータの近く（ご近所）にある、すでに対象が分かっているデータをk個集めて、一番多かったグループに所属させる」という手法
日常：「類は友を呼ぶ」や「郷に入っては郷に従う」という言葉

#・SVM
意味：データを2つのグループに分類するときに、「グループ間の隙間（マージン）が一番広くなるような、最高の境界線（超平面）」を見つけ出す手法
日常：「対立する2つのグループの間に、お互いが文句を言わないように、一番公平で広いド真ん中に境界線を引く」という行為

#・ノーフリーランチ
意味：「あらゆる問題に対して、常に完璧で一番優れている『万能のAIアルゴリズム』は存在しない」という定理
日常：「万能な道具や、完璧な選択肢なんて世の中にはない」
