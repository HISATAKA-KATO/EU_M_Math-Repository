```python
#8-5をRandomForestClassifierに置き換えよ。
#標準化ライブラリ
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier # ランダムフォレストをインポート

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# データの読み込み
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)

# デフォルトのランダムフォレストモデルの構築
forest = RandomForestClassifier(random_state=0)
forest.fit(X_train, y_train)

print(f"★デフォルト（ランダムフォレスト）★")
print(f"訓練セットのスコア: {forest.score(X_train, y_train):.3f}")
print(f"テストセットのスコア: {forest.score(X_test, y_test):.3f}")
```

```python
# パラメータを変更したモデル（木の深さを制限）
forest_trans = RandomForestClassifier(max_depth=4, random_state=0)
forest_trans.fit(X_train, y_train)

print(f"★変更後（ランダムフォレスト：深さ4）★")
print(f"訓練セットのスコア: {forest_trans.score(X_train, y_train):.3f}")
print(f"テストセットのスコア: {forest_trans.score(X_test, y_test):.3f}")
```
